# Análisis de Correlación para Factores de Demanda

**Problema**: Una empresa analiza cómo diferentes factores afectan la demanda de un producto:

| Factor      | Precipitación (mm) | Temperatura (°C) | Demanda (unidades) |
|-------------|---------------------|------------------|--------------------|
| **Día 1**   | 10                  | 25               | 300                |
| **Día 2**   | 20                  | 20               | 250                |
| **Día 3**   | 5                   | 30               | 400                |
| **Día 4**   | 15                  | 22               | 270                |
| **Día 5**   | 8                   | 28               | 350                |

**Preguntas**:

1. Calcula los coeficientes de correlación entre la demanda y los factores climáticos (precipitación y temperatura).  
2. Discute qué factor tiene mayor impacto en la demanda.  

**Fórmula para el Coeficiente de Correlación de Pearson**:

$$
r_{X, Y} = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum (X_i - \bar{X})^2 \cdot \sum (Y_i - \bar{Y})^2}}
$$

Donde:  
- $X, Y$: Variables de interés (e.g., demanda y precipitación, demanda y temperatura).  
- $\bar{X}, \bar{Y}$: Promedios de las variables $X$ e $Y$.  


# Cálculo Matemático Corregido: Análisis de Correlación

## Fórmula del Coeficiente de Correlación

El coeficiente de correlación de Pearson se calcula como:

$$
r_{X, Y} = \frac{\sum (X_i - \bar{X})(Y_i - \bar{Y})}{\sqrt{\sum (X_i - \bar{X})^2 \cdot \sum (Y_i - \bar{Y})^2}}
$$

---

## Paso 1: Datos del Problema

| Día    | $X_1$ (Precipitación) | $X_2$ (Temperatura) | $Y$ (Demanda) |
|--------|------------------------|---------------------|---------------|
| **1**  | 10                     | 25                  | 300           |
| **2**  | 20                     | 20                  | 250           |
| **3**  | 5                      | 30                  | 400           |
| **4**  | 15                     | 22                  | 270           |
| **5**  | 8                      | 28                  | 350           |

---

## Paso 2: Promedios

Calculamos los promedios:

1. **Precipitación** $(X_1)$:
   $$
   \bar{X_1} = \frac{10 + 20 + 5 + 15 + 8}{5} = 11.6
   $$

2. **Temperatura** $(X_2)$:
   $$
   \bar{X_2} = \frac{25 + 20 + 30 + 22 + 28}{5} = 25
   $$

3. **Demanda** $(Y)$:
   $$
   \bar{Y} = \frac{300 + 250 + 400 + 270 + 350}{5} = 314
   $$

---

## Paso 3: Desviaciones y Productos

### Precipitación $(X_1)$ vs Demanda $(Y)$:

| Día    | $X_1 - \bar{X_1}$ | $Y - \bar{Y}$ | $(X_1 - \bar{X_1})^2$ | $(Y - \bar{Y})^2$ | $(X_1 - \bar{X_1})(Y - \bar{Y})$ |
|--------|--------------------|---------------|------------------------|-------------------|---------------------------------|
| **1**  | $10 - 11.6 = -1.6$ | $300 - 314 = -14$ | $(-1.6)^2 = 2.56$   | $(-14)^2 = 196$  | $(-1.6)(-14) = 22.4$          |
| **2**  | $20 - 11.6 = 8.4$  | $250 - 314 = -64$ | $(8.4)^2 = 70.56$   | $(-64)^2 = 4096$ | $(8.4)(-64) = -537.6$         |
| **3**  | $5 - 11.6 = -6.6$  | $400 - 314 = 86$  | $(-6.6)^2 = 43.56$  | $(86)^2 = 7396$  | $(-6.6)(86) = -567.6$         |
| **4**  | $15 - 11.6 = 3.4$  | $270 - 314 = -44$ | $(3.4)^2 = 11.56$   | $(-44)^2 = 1936$ | $(3.4)(-44) = -149.6$         |
| **5**  | $8 - 11.6 = -3.6$  | $350 - 314 = 36$  | $(-3.6)^2 = 12.96$  | $(36)^2 = 1296$  | $(-3.6)(36) = -129.6$         |

Sumas:
$$
\sum (X_1 - \bar{X_1})^2 = 141.2, \quad \sum (Y - \bar{Y})^2 = 14820, \quad \sum (X_1 - \bar{X_1})(Y - \bar{Y}) = -1361.6
$$

### Temperatura $(X_2)$ vs Demanda $(Y)$:

| Día    | $X_2 - \bar{X_2}$ | $Y - \bar{Y}$ | $(X_2 - \bar{X_2})^2$ | $(Y - \bar{Y})^2$ | $(X_2 - \bar{X_2})(Y - \bar{Y})$ |
|--------|--------------------|---------------|------------------------|-------------------|---------------------------------|
| **1**  | $25 - 25 = 0$      | $-14$         | $(0)^2 = 0$           | $196$             | $(0)(-14) = 0$                |
| **2**  | $20 - 25 = -5$     | $-64$         | $(-5)^2 = 25$         | $4096$            | $(-5)(-64) = 320$             |
| **3**  | $30 - 25 = 5$      | $86$          | $(5)^2 = 25$          | $7396$            | $(5)(86) = 430$               |
| **4**  | $22 - 25 = -3$     | $-44$         | $(-3)^2 = 9$          | $1936$            | $(-3)(-44) = 132$             |
| **5**  | $28 - 25 = 3$      | $36$          | $(3)^2 = 9$           | $1296$            | $(3)(36) = 108$               |

Sumas:
$$
\sum (X_2 - \bar{X_2})^2 = 68, \quad \sum (X_2 - \bar{X_2})(Y - \bar{Y}) = 990
$$

---

## Paso 4: Cálculo de Correlaciones

1. **Precipitación y Demanda**:
   $$
   r_{X_1, Y} = \frac{\sum (X_1 - \bar{X_1})(Y - \bar{Y})}{\sqrt{\sum (X_1 - \bar{X_1})^2 \cdot \sum (Y - \bar{Y})^2}}
   $$
   $$
   r_{X_1, Y} = \frac{-1361.6}{\sqrt{141.2 \cdot 14820}} \approx -0.94
   $$

2. **Temperatura y Demanda**:
   $$
   r_{X_2, Y} = \frac{\sum (X_2 - \bar{X_2})(Y - \bar{Y})}{\sqrt{\sum (X_2 - \bar{X_2})^2 \cdot \sum (Y - \bar{Y})^2}}
   $$
   $$
   r_{X_2, Y} = \frac{990}{\sqrt{68 \cdot 14820}} \approx 0.98
   $$

---

## Conclusión

1. La correlación entre **precipitación y demanda** es \$( r \approx -0.94 \$), indicando una correlación negativa fuerte.  
2. La correlación entre **temperatura y demanda** es \$( r \approx 0.98 \$), indicando una correlación positiva fuerte.  
3. La **temperatura** tiene un impacto mayor en la demanda que la **precipitación**.


```python
import numpy as np
import matplotlib.pyplot as plt

# Datos del problema
precipitacion = np.array([10, 20, 5, 15, 8])  # Precipitación (mm)
temperatura = np.array([25, 20, 30, 22, 28])  # Temperatura (°C)
demanda = np.array([300, 250, 400, 270, 350])  # Demanda (unidades)

# Calcular coeficientes de correlación
correlacion_precipitacion_demanda = np.corrcoef(precipitacion, demanda)[0, 1]
correlacion_temperatura_demanda = np.corrcoef(temperatura, demanda)[0, 1]

# Crear gráficas
plt.figure(figsize=(14, 6))

# Gráfica 1: Precipitación vs Demanda
plt.subplot(1, 2, 1)
plt.scatter(precipitacion, demanda, color='blue', label=f'r = {correlacion_precipitacion_demanda:.2f}')
plt.plot(np.unique(precipitacion), np.poly1d(np.polyfit(precipitacion, demanda, 1))(np.unique(precipitacion)), color='red')
plt.xlabel("Precipitación (mm)")
plt.ylabel("Demanda (unidades)")
plt.title("Precipitación vs Demanda")
plt.legend()
plt.grid(True)

# Gráfica 2: Temperatura vs Demanda
plt.subplot(1, 2, 2)
plt.scatter(temperatura, demanda, color='green', label=f'r = {correlacion_temperatura_demanda:.2f}')
plt.plot(np.unique(temperatura), np.poly1d(np.polyfit(temperatura, demanda, 1))(np.unique(temperatura)), color='red')
plt.xlabel("Temperatura (°C)")
plt.ylabel("Demanda (unidades)")
plt.title("Temperatura vs Demanda")
plt.legend()
plt.grid(True)

# Mostrar las gráficas
plt.tight_layout()
plt.show()

# Resultados de correlaciones
correlacion_precipitacion_demanda, correlacion_temperatura_demanda
```
